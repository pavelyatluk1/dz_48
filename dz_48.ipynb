Цей датасет містить чотири колонки, що відповідають:  - помірно зміщеним вправо даним (Moderate Positive Skew), - сильно зміщеним вправо даним (Highly Positive Skew), - помірно зміщеним вліво даним (Moderate Negative Skew), - сильно зміщеним вліво даним (Highly Negative Skew), - Оцініть результати трансформацій через квантильно-квантильні графіки


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
```

```python
# завантажуємо датасет
df = pd.read_csv('Data_to_Transform.csv')
df.head()
```

### Трансформація колонки 'Moderate Positive Skew'

```python
# гістограма розподілу
sns.histplot(df['Moderate Positive Skew'])
fig = plt.figure()
```

```python
# логарифмічна трансформація
df_log = np.log(df['Moderate Positive Skew'])
sns.histplot(df_log)
fig = plt.figure()
```

```python
# трансфрмація кубічним коренем
df_cbrt = np.cbrt(df['Moderate Positive Skew'])
sns.histplot(df_cbrt)
fig = plt.figure()
```

```python
# трансформація Бокса-Кокса
from sklearn.preprocessing import PowerTransformer
data = df['Moderate Positive Skew'].values.reshape(-1, 1)
```

```python
pt = PowerTransformer(method='box-cox')
```

```python
df_boxcox = pt.fit_transform(data)
sns.histplot(df_boxcox)
fig = plt.figure()
```

### Оцінка нормальності розподулу Q-Q plot

```python
df = np.int64(df['Moderate Positive Skew'])
fig = sm.qqplot(df, line='45')
plt.show()
```

```python
# оцінка нормальності перетоворення кубічним коронем
fig = sm.qqplot(df_cbrt, line='45')
plt.show()
```

```python
# оцінка логарифмічного перетворення
fig = sm.qqplot(df_log, line='45')
plt.show()
```

```python
# оцінка перетворення Бокса-Кокса
fig = sm.qqplot(df_boxcox, line='45')
plt.show()
```
